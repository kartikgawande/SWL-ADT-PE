## let's look into other ids that aint ending with zero but still having multiple table names
like 4.12

```python
attr[attr['Table_id']=='4.12']
```

## here it's clearly a typing error. the years are not completely written
let's rectify that

```python
wrong_table_names = attr[attr['Table_id']=='4.12']['Table_name'].unique().tolist()
wrong_table_names
```

to automate this editing let's keep the most lengthy table_name for all the table_id entries.

```python
def max_len_string(stringList: list[str]):
    max=-1
    for string in stringList:
        if len(string)>max:
            max=len(string)
            max_string=string
    return max_string
```

```python
max_len_string(wrong_table_names)
```

```python
attr.loc[attr['Table_id']=='4.12','Table_name']=max_len_string(wrong_table_names)
```

```python
attr[attr['Table_id']=='4.12']
```

```python
table_name_count = attr.groupby('Table_id')['Table_name'].nunique().reset_index()
table_name_count
multiple_table_name = table_name_count[table_name_count['Table_name']>1]
multiple_table_name
```

```python
multiple_table_name.shape
```

let's do this for all the table_ids

```python
faulty_table_ids_list = multiple_table_name['Table_id'].to_list()
faulty_table_ids_list
```

```python
for table_id in faulty_table_ids_list:
    wrong_table_names = attr[attr['Table_id']==table_id]['Table_name'].unique().tolist()
    attr.loc[attr['Table_id']==table_id,'Table_name']=max_len_string(wrong_table_names)
```

```python
table_name_count = attr.groupby('Table_id')['Table_name'].nunique().reset_index()
table_name_count
multiple_table_name = table_name_count[table_name_count['Table_name']>1]
multiple_table_name
```

now all the table_name descrepancies have been removed successfully
